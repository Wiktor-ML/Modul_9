Pogrupuj tabelę w taki sposób, aby otrzymać średni przychód (revenue) oraz średni budżet (budget) w danym roku dla filmów opublikowanych od 2010 (włącznie) do 2016 roku (włącznie). Następnie na tej podstawie stwórz wykres, w którym średnie przychody są wykresem kolumnowym, a średni budżet wykresem liniowym na tych samych osiach. Sformatuj odpowiednio oś X oraz oś Y. Dodaj tytuł wykresu, oraz legendę, która znajduje się w prawym górnym rogu płótna, lecz poza obszarem osi. Przykład wykresu widoczny poniżej

```python
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'])
```

```python
filtered_df_movies = df_movies[(df_movies['release_date'] >= '2010-01-01') & (df_movies['release_date'] <= '2016-12-31')]
```

```python
filtered_df_movies.head()
```

```python
filtered_df_movies = filtered_df_movies.assign(year=filtered_df_movies['release_date'].dt.year)
```

```python
filtered_df_movies.head()
```

```python
mean_budgets_by_year = filtered_df_movies.groupby('year')['budget'].mean()
```

```python
print(mean_budgets_by_year)
```

```python
mean_revenue_by_year = filtered_df_movies.groupby('year')['revenue'].mean()
```

```python
print(mean_revenue_by_year)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# Create a plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the revenue as bars
years = mean_revenue_by_year.index # use the years from the index of mean_revenue
mean_revenue_by_year_plt = ax.bar(years, mean_revenue_by_year, label="revenue")

mean_budgets_by_year_plt = ax.plot(years, mean_budgets_by_year, color='red', linestyle='-', label="budget")

def million(x, pos):
        return '{:2.1f}M'.format(x*1e-6)
    
# Formatting the axes
formatter = plt.FuncFormatter(million)
ax.yaxis.set_major_formatter(formatter)

# Add a legend
ax.legend(loc=(1.02,0.91))


plt.title('Sredni przychod i budzet filmu w latach 2010-2016')
ax.tick_params(axis='x', rotation=0)
ax.set_xlabel("")

# Ensure the labels aren't obscured by the plot
plt.tight_layout()
plt.show()
```
